```python
def color_thresholds(image,sthresh=(0,255),vthresh=(0,255)):
    hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
    s_channel = hls[:,:,2]
    s_binary = np.zeros_like(s_channel)
    s_binary[(s_channel > sthresh[0]) & (s_channel <= sthresh[1])] = 1
    
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    v_channel = hsv[:,:,2]
    v_binary = np.zeros_like(v_channel)
    v_binary[(v_channel > vthresh[0]) & (v_channel <= vthresh[1])] = 1
    
    sv_binary = np.zeros_like(s_channel)
    sv_binary[(s_binary == 1) & ( v_binary == 1)] = 1
    return sv_binary

image = mpimg.imread('test_images/test6.jpg')
sv_binary_image = color_thresholds(image,(100,255),(50,255))
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(image,cmap='gray')
ax1.set_title('gray Image', fontsize=30)

ax2.imshow(sv_binary_image,cmap='gray')
ax2.set_title('SV binary Image', fontsize=30)
```
